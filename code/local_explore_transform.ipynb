```python
import pandas as pd
import os
```

#### import files local

```python
base_dir = os.getcwd()
```

```python
# csv file path
csv_songs_path = os.path.join(base_dir, "..", "data", "songs.csv")
csv_users_path = os.path.join(base_dir, "..", "data", "users.csv")
csv_stream_folder = os.path.join(base_dir, "..", "data")

df_songs = pd.read_csv(csv_songs_path)
df_users =pd.read_csv(csv_users_path)
df1 = pd.read_csv(f"{csv_stream_folder}\\streams1.csv")
df2 = pd.read_csv(f"{csv_stream_folder}\\streams2.csv")
df3 = pd.read_csv(f"{csv_stream_folder}\\streams3.csv")
df_streams = pd.concat([df1, df2, df3], ignore_index=True)
```

#### Cleaning

```python
# change data type
df_streams['listen_time'] = pd.to_datetime(df_streams['listen_time'])
df_users['created_at'] = pd.to_datetime(df_users['created_at'])
```

```python
df_users.info(), df_streams.info()
```

```python
# handle missing values if exists
def check_handling_nan(df, name):
    any_nan = df.isna().any().any()
    if any_nan:
        # handle nan in that df
        for col in df.columns:
            if df[col].isna().any():
                print(f"Handle missing values on df {name}, column {col}")
                if df[col].dtype == 'object':
                    df[col] = df[col].fillna('Unknown')
                elif pd.api.types.is_datetime64_any_dtype(df[col]):
                    df[col] = df[col].fillna(pd.to_datetime('9999-12-31'))
                else:
                    df[col] = df[col].fillna(0)
    return df
```

```python
df_streams = check_handling_nan(df_streams, "streams")
print(f"Missing values:\n{df_streams.isna().sum()}")

df_users = check_handling_nan(df_users, "users")
print(f"Missing values:\n{df_users.isna().sum()}")

df_songs = check_handling_nan(df_songs, "songs")
print(f"Missing values:\n{df_songs.isna().sum()}")
```

```python
def check_handling_duplicate(df, id_column, name):
    if df.duplicated(subset=id_column).any():
        print(f"Found duplicate on {name}, drop duplicate...")
        df = df.drop_duplicates(subset=id_column, keep='first')
    print(f"No duplicate on {name}")
    return df
```

```python
df_songs = check_handling_duplicate(df_songs, id_column='track_id', name="songs")
df_users = check_handling_duplicate(df_users, id_column='user_id', name="users")
df_streams = check_handling_duplicate(df_streams, id_column=['user_id', 'track_id', 'listen_time'], name="streams")
```

#### Join

```python
# main table for analysis
streams_full_clean_analysis = df_streams.merge(df_users, on="user_id", how="inner") \
                                        .merge(df_songs, on="track_id", how="inner")

# temp table for completeness
streams_full_archieve = df_streams.merge(df_users, on="user_id", how="left") \
                                  .merge(df_songs, on="track_id", how="left")
```

```python
missing_users_df = streams_full_archieve[streams_full_archieve['user_id'].isna()]
missing_songs_df = streams_full_archieve[streams_full_archieve['track_id'].isna()]

missing_users_count = len(missing_users_df)
missing_songs_count = len(missing_songs_df)

print(f"[INFO] Missing users: {missing_users_count}")
print(f"[INFO] Missing songs: {missing_songs_count}")

# save hanya jika ada data user atau song yang hilang
if missing_users_count > 0:
    missing_users_df.to_csv("audit_missing_users.csv", index=False)
    print("[INFO] Saved audit_missing_users.csv")

if missing_songs_count > 0:
    missing_songs_df.to_csv("audit_missing_songs.csv", index=False)
    print("[INFO] Saved audit_missing_songs.csv")
```

#### Genre-level KPIs

```python
genre_level_kpis = streams_full_clean_analysis.groupby("track_genre").agg(
    total_streams=("listen_time", "count"),
    avg_popularity=("popularity", "mean"),
    unique_listeners=("user_id", "nunique"),
    avg_duration=("duration_ms", "mean")    
).reset_index()
```

```python
genre_explicit_kpi = df_songs.groupby("track_genre").agg(
    total_track=("track_id", "count"),
    total_explicit_track=("explicit", "sum")
).reset_index()

genre_explicit_kpi['explicit_ratio'] = genre_explicit_kpi['total_explicit_track']/genre_explicit_kpi['total_track']
```

#### Hourly-level KPIs

```python
streams_full_clean_analysis['listen_hour'] = streams_full_clean_analysis['listen_time'].dt.hour
streams_full_clean_analysis.columns
```

```python
hourly_level_kpi = streams_full_clean_analysis.groupby(["listen_date", "listen_hour"]).agg(
    total_streams=("listen_time", "count"),
    unique_listeners=("user_id", "nunique"),
).reset_index()
hourly_level_kpi
```

```python
top_genre_hourly = streams_full_clean_analysis.groupby(["listen_date", "listen_hour", "track_genre"]).agg(
    total_streams=("track_genre", "count")
).reset_index() \
.sort_values(["listen_date", "listen_hour", "total_streams"], ascending=[True, True, False]) \
.drop_duplicates(subset=["listen_hour"], keep="first")
top_genre_hourly
```

```python
top_artist_hourly = streams_full_clean_analysis.groupby(["listen_date", "listen_hour", "artists"]).agg(
    total_streams=("artists", "count")
).reset_index() \
.sort_values(["listen_date", "listen_hour", "total_streams"], ascending=[True, True, False]) \
.drop_duplicates(subset=["listen_hour"], keep="first")
top_artist_hourly
```

```python
bins = list(range(10, 101 + 1, 10))  # [11, 21, 31, ..., 101]
labels = [f"{i}-{i+9}" for i in bins[:-1]]  # ["11-20", "21-30", ..., "91-100"]
streams_full_clean_analysis['age_group'] = pd.cut(
    streams_full_clean_analysis['user_age'],
    bins=bins,
    labels=labels,
    right=False,
    include_lowest=True
)
streams_full_clean_analysis.head()
```

```python
mos_engaged_age_group_hourly = streams_full_clean_analysis.groupby(["listen_date", "listen_hour", "age_group"]).agg(
    total_streams=("age_group", "count")
).reset_index() \
.sort_values(["listen_date", "listen_hour", "total_streams"], ascending=[True, True, False]) \
.drop_duplicates(subset=["listen_hour"], keep="first")
mos_engaged_age_group_hourly
```

```python
hourly_level_kpi = hourly_level_kpi.merge(top_genre_hourly[["listen_date", "listen_hour", "track_genre"]], on=["listen_date", "listen_hour"], how="left")
```

```python
hourly_level_kpi = hourly_level_kpi.merge(top_artist_hourly[["listen_date", "listen_hour", "artists"]], on=["listen_date", "listen_hour"], how="left")
```

```python
hourly_level_kpi = hourly_level_kpi.merge(mos_engaged_age_group_hourly[["listen_date", "listen_hour", "age_group"]], on=["listen_date", "listen_hour"], how="left")
```
